### rune çš„å¯¦éš›æ‡‰ç”¨

````go
package main

import (
    "fmt"
    "unicode"
    "unicode/utf8"
)

func main() {
    // éŒ¯èª¤çš„å­—ä¸²è™•ç†æ–¹å¼ (æŒ‰ byte ç´¢å¼•)
    text := "Goèªè¨€"
    fmt.Printf("åŸå§‹å­—ä¸²: %s\n", text)
    
    // éŒ¯èª¤ï¼šæŒ‰ byte ç´¢å¼•æœƒåˆ‡æ–· Unicode å­—å…ƒ
    fmt.Printf("éŒ¯èª¤çš„åˆ‡ç‰‡ text[0:3]: %s\n", text[0:3]) // "Goèª" çš„ä¸€éƒ¨åˆ†
    
    // æ­£ç¢ºçš„å­—ä¸²è™•ç†æ–¹å¼ (è½‰æ›ç‚º rune slice)
    runes := []rune(text)
    fmt.Printf("è½‰æ›ç‚º rune slice: %v\n", runes)
    fmt.Printf("æ­£ç¢ºçš„åˆ‡ç‰‡ runes[0:3]: %s\n", string(runes[0:3])) // "Goèª"
    
    // å­—å…ƒåˆ†é¡
    fmt.Printf("\nå­—å…ƒåˆ†æ:\n")
    for i, r := range text {
        fmt.Printf("ä½ç½® %d: %c (Unicode: U+%04X)\n", i, r, r)
        fmt.Printf("  æ˜¯å­—æ¯: %v\n", unicode.IsLetter(r))
        fmt.Printf("  æ˜¯æ•¸å­—: %v\n", unicode.IsDigit(r))
        fmt.Printf("  æ˜¯ä¸­æ–‡: %v\n", unicode.Is(unicode.Han, r))
        fmt.Println()
    }
    
    // å­—ä¸²åè½‰ (æ­£ç¢ºè™•ç† Unicode)
    original := "Hello ä¸–ç•Œ! ğŸŒ"
    reversed := reverseString(original)
    fmt.Printf("åŸå§‹: %s\n", original)
    fmt.Printf("åè½‰: %s\n", reversed)
    
    // UTF-8 ç·¨ç¢¼è³‡è¨Š
    fmt.Printf("\nUTF-8 ç·¨ç¢¼è³‡è¨Š:\n")
    for _, r := range "Aä¸­ğŸš€" {
        size := utf8.RuneLen(r)
        fmt.Printf("å­—å…ƒ %c: %d bytes\n", r, size)
    }
}

// æ­£ç¢ºçš„å­—ä¸²åè½‰å‡½å¼ (è™•ç† Unicode)
func reverseString(s string) string {
    runes := []rune(s)
    for i, j := 0, len(runes)-1; i < j; i, j = i+1, j-1 {
        runes[i], runes[j] = runes[j], runes[i]
    }
    return string(runes)
}
````
